+ 결측치 확인 및 처리

```python
# 결측치 확인 및 처리
def na_preprocess(df, anomaly_per, na):

    if na is True:

        #Column별 결측치 n% 이상 있을 경우 제외
        remove_v1 = round(df.isnull().sum() / len(df)*100, 2)
        tmp_df = df[remove_v1[remove_v1 < anomaly_per].index]

        #Row별 결측치 n% 이상 있을 경우 제외
        idx1 = len(tmp_df.columns) * 0.7
        tmp_df.dropna(thresh=idx1, axis=0)

    else:

        #보강
        tmp_df = df.fillna(0)

    #print('결측치 처리')
    return tmp_df

if na_pre:
    data = na_preprocess(df=data, anomaly_per = _anomaly_ratio, na = na_pre)
```

# **2. 분석 알고리즘을 위한 페이지**

+ '추천 알고리즘' 스크롤을 통해 알고리즘 선택
+ 본 사업에서는 어떤 알고리즘을 선택해도 같은 결과가 나오도록 함

```python
def calculate_similarity(df, latent_features=50):
    pt = pd.pivot_table(index='USER_ID', columns='ITEM_ID', values='EVENT_TYPE', data=df, fill_value=0) #interaction데이터 pivot table 생성
    pt_m = pt.to_numpy()                                     # pivot table 생성
    user_ratings_mean = np.mean(pt_m, axis = 1)              # 사용자별 rating 평균 구하기
    df_demeaned = pt_m - user_ratings_mean.reshape(-1, 1)    # pivot table에서 rating 평균 빼기
    U, sigma, Vt = svds(df_demeaned, latent_features)        # SVD 분해  
    sigma = np.diag(sigma)                        
    all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
    preds_df = pd.DataFrame(all_user_predicted_ratings, index=pt.index, columns = pt.columns)
    return preds_df
```

```python
ml_preds_df = calculate_similarity(interaction)
ml_preds_df.head(5)
```

```python
def intersect(a, b):
    return list(set(a) & set(b))

#아이템별 precision@k와 recall@k의 값을 알려줌
def precision_recall(predictions_df, item_df, interaction_df, USER_ID, num_recommendations=_num_recommendations):
   
    sorted_user_predictions = predictions_df.iloc[int(USER_ID-1), :].sort_values(ascending=False).head(num_recommendations)

    user_data = interaction_df[interaction_df.USER_ID==USER_ID]
    user_item = item_df[item_df.ITEM_ID.isin(user_data.ITEM_ID.tolist())]

    recommended_items = sorted_user_predictions.index.tolist()
    user_item_ = user_item.ITEM_ID.tolist()

    precision = len(intersect(user_item_, recommended_items))/ len(recommended_items)
    recall = len(intersect(user_item_, recommended_items)) / len(user_item_)
    
    return precision, recall

def perf_metric(preds_df, item, interaction, num_recommendations=_num_recommendations):
    dict_list = []
    for user_id in interaction.USER_ID.unique():
        precision, recall = precision_recall(preds_df, item, interaction, USER_ID = user_id, num_recommendations=num_recommendations)
        dict_ = {}
        dict_.update({"userID" : user_id})
        dict_.update({f"precision@{num_recommendations}" : precision})
        dict_.update({f"recall@{num_recommendations}" : recall})
        dict_list.append(dict_)
    accuracy_df = pd.DataFrame(dict_list)
    return accuracy_df
```

```python
precision, recall = precision_recall(ml_preds_df, item, interaction, USER_ID = 1, num_recommendations=_num_recommendations)
```

```python
accuracy_df = perf_metric(ml_preds_df, item, interaction, num_recommendations=_num_recommendations)
```

```python
#USER_ID 가 1일 때 추천 아이템 선정
USER_ID=45

def make_each_recommendations(ml_preds_df, interaction, item, USER_ID):
    #USER_ID 별 상위 ITEM 추천 목록 생성
    sorted_user_predictions = ml_preds_df.iloc[int(USER_ID-1), :].sort_values(ascending=False)
    sorted_user_predictions = pd.DataFrame(sorted_user_predictions.reset_index())
    sorted_user_predictions.columns = ['ITEM_ID', 'Predictions']

    #이미 USER_ID가 선정했던 ITEM 확인
    user_data = interaction[interaction.USER_ID==USER_ID]
    user_item = item[item.ITEM_ID.isin(user_data.ITEM_ID.tolist())]
    print('User {0} has already viewed or buyed {1} items'.format(USER_ID, user_item.shape[0]))

    #상위 추천 ITEM중 이미 USER_ID가 선정했던 ITEM 제외 후 추천 목록 추출
    recommended_items = sorted_user_predictions.index.tolist()
    recommendations = item[~item['ITEM_ID'].isin(user_item['ITEM_ID'])].merge(sorted_user_predictions, how='inner',on='ITEM_ID')
    recommendations = recommendations.sort_values('Predictions', ascending=False).head(_num_recommendations)

    return user_item.reset_index(drop=True), recommendations.reset_index(drop=True)

user_item, recommendations = make_each_recommendations(ml_preds_df, interaction, item, USER_ID)
```
