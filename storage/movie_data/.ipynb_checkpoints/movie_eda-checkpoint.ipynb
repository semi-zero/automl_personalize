```python
import pandas as pd
import json
import numpy as np
import time
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
items = pd.read_csv('items.csv')
item_cols = ['id', 'name', 'category', 'style', 'description','price']
items_df = items[item_cols]
items_df = items_df.rename(columns={'id':'ITEM_ID',
                              'name' : 'ITEM_NAME',
                              'category' :'CATEGORY_L1',
                              'style' : 'STYLE',
                              'description' : 'ITEM_DESCRIPTION',
                              'price' : 'PRICE',
                             })
print(items_df.shape)
items_df.head()
```

```python
users = pd.read_csv('users.csv')
user_cols = ['id', 'username', 'age', 'gender']
users_df = users[user_cols]
users_df = users_df.rename(columns={'id':'USER_ID',
                              'username' : 'USER_NAME',
                              'age' :'AGE',
                              'gender' : 'GENDER',                              
                             })

users_df['USER_ID'] = users_df['USER_ID'].astype('object')
print(users_df.shape)
print(users_df.dtypes)
users_df.head()
```

```python
its = pd.read_csv('interactions.csv')
select_event_types = ['ProductViewed','OrderCompleted']

its_df = its[its['EVENT_TYPE'].isin(select_event_types)]
its_df['EVENT_TYPE'] = its_df['EVENT_TYPE'].replace(['ProductViewed'],'View')    
its_df['EVENT_TYPE'] = its_df['EVENT_TYPE'].replace(['OrderCompleted'],'Purchase')

its_df['USER_ID'] = its_df['USER_ID'].astype('object')
print(its_df.shape)
print(its_df.dtypes)
its_df.head(3)
```

```python
merge1 = pd.merge(its_df, users_df, on='USER_ID', how='left')
df = pd.merge(merge1, items_df, on='ITEM_ID', how='left')
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'], unit='s')
```

```python
print(df.shape)
df.head()
```

```python
df.to_csv('interaction_data.csv', index=False)
```

```python
pd.options.display.max_rows = 5
def split_holdout(data, pct):
    df = data.copy()
    # Rank per each subgroup, 'USER_ID'
    ranks = df.groupby('USER_ID').TIMESTAMP.rank(pct=True, method='first')
    df = df.join((ranks> pct).to_frame('holdout'))
    
    holdout = df[df['holdout']].drop('holdout', axis=1)
    train = df[~df['holdout']].drop('holdout', axis=1)    
    
    return train, holdout

df_warm_train, df_warm_holdout = split_holdout(df, pct=0.9)
```
